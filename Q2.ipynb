period of second signal : 32 = 32(points on DFT axis), and
period of third signal : 34 > 32(points on DFT axis)

Spectrum Leakage is there in third signal because it has greater period than the points for which DFT is calculated. So, we can't estimate the exact frequency of the input signal by looking as we get more than one non-zero frequency components.

So, to avoid spectrum leakage for M point DFT of signal with period N, M must be greater than or equal to N (M>=N).
