Q6. For the LTI systems described by the following difference equation, generate the frequency response. Obtain the pole zero plot. Comment on the type of response (stable/causal etc)

ùë¶[ùëõ]+1.3ùë¶[ùëõ‚àí1]+1.29ùë¶[ùëõ‚àí2]+0.52ùë¶[ùëõ‚àí3]+0.16ùë¶[ùëõ‚àí4]=ùë•[ùëõ]‚àí0.9ùë•[ùëõ‚àí1] 
ùë¶[ùëõ]‚àí1.3ùë¶[ùëõ‚àí1]+1.29ùë¶[ùëõ‚àí2]‚àí0.52ùë¶[ùëõ‚àí3]+0.16ùë¶[ùëõ‚àí4]=ùë•[ùëõ]+0.9ùë•[ùëõ‚àí1]

```python
from scipy import signal
import numpy as np 
import matplotlib.pyplot as plt

# Question1

b = np.array([1, -0.9, 0, 0, 0])
a = np.array([1, 1.3, 1.29, 0.52, 0.16])

print("Zeros and poles are respectively:")
print(signal.tf2zpk(b, a))

w, h = signal.freqz(b,a)
fig, ax1 = plt.subplots()
ax1.set_title('frequency response')
ax1.plot(w, 20 * np.log10(abs(h)), 'b')
ax1.set_ylabel('Amplitude (in dB)', color='b')
ax1.set_xlabel('Frequency (in rad/sample)')
ax2 = ax1.twinx()
angles = np.unwrap(np.angle(h))
ax2.plot(w, angles, 'r')
ax2.set_ylabel('Angle (radians)', color='r')
ax2.grid()
ax2.axis('tight')
plt.show()

```

```python
#Question2 

b1 = np.array([1, 0.9, 0, 0, 0])
a1 = np.array([1, -1.3, 1.29, -0.52, 0.16])

print("Zero's and poles are respectively:")
print(signal.tf2zpk(b1, a1))

w, h = signal.freqz(b1,a1)
fig, ax1 = plt.subplots()
ax1.set_title('frequency response')
ax1.plot(w, 20 * np.log10(abs(h)), 'b')
ax1.set_ylabel('Amplitude (dB)', color='b')
ax1.set_xlabel('Frequency (rad/sample)')
ax2 = ax1.twinx()
angles = np.unwrap(np.angle(h))
ax2.plot(w, angles, 'r')
ax2.set_ylabel('Angle (radians)', color='r')
ax2.grid()
ax2.axis('tight')
plt.show()
```
