Consider a round theatre where an orchestra is in the middle of two concentric circles and the walls on one half side are at a radial distances of  17.15𝑚  (inner circle) and  34.3𝑚  (outer circle) on the other side from the orchestra. The speed of sound is  343𝑚/𝑠 . Assume that the recorded signal is the sum of the original signal and the attenuated echoes from the two walls and is given by  𝑟[𝑛]=𝑦[𝑛]+0.8𝑦[𝑛−𝑁1]+0.6𝑦[𝑛−𝑁2]  where,  𝑁1  is the delay caused by the closest wall and N2 is the delay caused by the farther wall. The recorder is in the centre of the theatre. Take any audio signal available and generate  𝑟[𝑛]  and listen to both the original and the echoed signal. Write your observations and upload the original and echoed signal.

```python
import librosa
import IPython.display as display
import numpy as np

org_data, fs = librosa.load('orchestra_sample.wav', sr=20000)
display.Audio(org_data,rate = fs) #original_signal
```

```python
#distanced travelled by signal to hit the closest wall and then to reach the recorder is 17.15 *2   =  34.3
#Given sound speed = 343m/s
#N1 will be 34.3/343 seconds
N1 = 34.3/343
N1 = N1*20000

#distanced travelled by signal to hit the closest wall and then to reach the recorder is 34.3 *2  =  68.6
N2 = 68.6/343
N2 = N2*20000 #gives the number of samples for required delay

delay1 = org_data*0.8
delay2 = org_data*0.6

N1 = np.int(np.ceil(N1))
delay1_zeros = np.zeros(N1)
delay1 = np.concatenate((delay1_zeros,delay1))
N2 = np.int(np.ceil(N2))
delay2_zeros = np.zeros(N2)
delay2 = np.concatenate((delay2_zeros,delay2))

x = np.zeros((delay2.shape[0] - delay1.shape[0]))
delay1 = np.concatenate((delay1,x))
modified_org_data = np.concatenate((org_data,delay2_zeros))
final_data = modified_org_data + delay1 + delay2
display.Audio(final_data,rate = fs) #final_signal
```
