Q1. Compute the DTFT (magnitude and phase) of the following (use <i>scipy.signal.freqz</i>). Plot from $\omega = −2\pi$ to $2\pi$. Observe the periodicty, symmetries and relations between the spectra.

<ol>
  <li>$r[n] = u[n] − u[n − 5]$  </li>
    <li>$(−1)^nr[n]$ </li>
</ol>

```python
#Q1.1)

import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

x = np.zeros([10000])
y = [1,1,1,1,1]

y = np.concatenate((y, x))

w, h = signal.freqz(y, whole = True, worN = 1024)

a = w[: :-1]
a = -1 * a

w = np.concatenate((a,w))
h = np.concatenate((h,h))

fig = plt.figure()
plt.title('frequency response')

amp = fig.add_subplot(311)
plt.plot(w, 20 * np.log10(abs(h)))
plt.ylabel('Amplitude [dB]')
plt.xlabel('Frequency [rad/sample]')
plt.grid()

phs = fig.add_subplot(313)
angles = np.unwrap(np.angle(h))
plt.plot(w, angles)
plt.ylabel('Angle (radians)')
plt.xlabel('Frequency [rad/sample]')
plt.grid()

plt.show()
```

```python
#Q1.2)

import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

x = np.zeros([10000])
y = [1,1,1,1,1]
y = np.concatenate((y, x))
z = np.arange(10005)
c = np.power(-1, z)
c = c * y

w, h = signal.freqz(c, whole = True, worN = 1024)

a = w[: :-1]
a = -1 * a

w = np.concatenate((a, w))
h = np.concatenate((h, h))

fig = plt.figure()
plt.title('frequency response')
ax1 = fig.add_subplot(311)
plt.plot(w, 20 * np.log10(abs(h)))
plt.ylabel('Amplitude [dB]')
plt.xlabel('Frequency [rad/sample]')
plt.grid()

ax2 = fig.add_subplot(313)
angles = np.unwrap(np.angle(h))
plt.plot(w, angles)
plt.ylabel('Angle (radians)')
plt.xlabel('Frequency [rad/sample]')
plt.grid()

plt.show()
```
