Q8. Generate the following signals  洧논1[洧녵]=洧녫洧녶洧(洧랢洧녵/15),  洧논2[洧녵]=洧녫洧녶洧(3洧랢洧녵/15)  for n = 0 to 31. Plot the magnitude of 64 point DFT of the two signals  洧녦1[洧녲]  and  洧녦2[洧녲]  (with zero padding). Now concatenate the two signals to form a 64 length sequence  洧논[洧녵] . Plot the 64 point DFT magnitude  洧녦[洧녲]  of  洧논[洧녵] . Also find the sum of two signals and plot the 64 point DFT of the result. Write down your observations.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft,ifft

n = np.arange(32)
a = np.arange(64)

#Generating x1 and its DFT
x1 = np.cos(np.pi*n/15)
x1_padded = np.pad(x1,(0,32))
print('x1:')
print(x1_padded)
plt.stem(a,x1_padded,use_line_collection = True)
plt.title("x1")
plt.show()

x1_dft = fft(x1,64)
plt.stem(a,abs(x1_dft), use_line_collection = True)
plt.title("64-point DFT of x1")
plt.show()
```

```python
#Generating x2 and its DFT
x2 = np.cos(3*np.pi*n/15)
x2_padded = np.pad(x2,(0,32))
print('x2:')
print(x2_padded)
plt.stem(a,x2_padded,use_line_collection = True)
plt.title("x2")
plt.show()

x2_dft = fft(x2,64)
plt.stem(a,abs(x2_dft), use_line_collection = True)
plt.title("64-point DFT of x2")
plt.show()
```

```python
#Generating x3 and its DFT
x3 = np.concatenate((x1,x2),axis= None)
print('x3:')
print(x3)
plt.stem(a,x3,use_line_collection = True)
plt.title("x3 = concatenate(x1,x2)")
plt.show()

x3_dft = fft(x3,64)
plt.stem(a,abs(x3_dft), use_line_collection = True)
plt.title("64-point DFT of x3")
plt.show()
```

```python
#Generating x4 and its DFT
x4 = np.add(x1_padded,x2_padded)
print('x4:')
print(x4)
plt.stem(a,x4,use_line_collection = True)
plt.title("x4 = x1 + x2 ")
plt.show()

x4_dft = fft(x4,64)
plt.stem(a,abs(x4_dft), use_line_collection = True)
plt.title("64-point DFT of x4")
plt.show()
```
