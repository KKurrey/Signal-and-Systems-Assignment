Q3. Record your own voice at a sampling rate of 16kHz for one of the following  vowels for 1s,  --  aa, ii, oo , E  ( आ   ई  ऊ  ए ) .   Please compute the DTFT and plot.   Write down your observations. 


```python
import numpy as np
import librosa
from scipy import signal
import matplotlib.pyplot as plt

y, s = librosa.load('Audio.wav', sr=16000, duration = 1)
w, h = signal.freqz(y,whole = True)

y = w[::-1]
y = -1*y

w = np.concatenate((y,w))
h = np.concatenate((h,h))
plt.plot(w,h)

plt.ylabel('DTFT')
plt.xlabel('Frequency [rad/sample]')
plt.grid()
```
