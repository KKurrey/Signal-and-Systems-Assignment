Q7. Consider the following low pass filters with impulse responses in the form of step function, triangular and sinc functions of length N, given by:
<ol>
  <li>$ h_S[n] = u[n] − u[n − (N − 1)]$ </li>
  <li> $ h_L[n] = tri(\frac{n}{N}) = \begin {cases} 1-\frac{|n|}{N}~~~,~~~|n|\le{N}\\0~~~~~~~~~~~~,~~~elsewhere \end{cases} $
  <li>$ h_L[n]$ = $sinc(\frac{n}{N})$ = $\frac{sin(πn/N)}{πn/N}; \,\, sinc(0) = 1$ </li>  
</ol>

Plot the magnitude response of each function for $N=4$ and $N=8$. How does the response of step, linear and windowed $sinc$ schemes compare with ideal low pass case

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
from scipy.fftpack import fft, fftshift

#Question1 for N= 4
h1 = np.zeros((1000))
h1a = np.concatenate((np.ones(3),h1))
w, h = signal.freqz(h1a,whole = True)
y = w[::-1]
y = -1*y
w = np.concatenate((y,w))
h = np.concatenate((h,h))
fig = plt.figure()
plt.title('frequency response of step function N = 4')
plt.plot(w, h)
plt.ylabel('H(e^jw)')
plt.xlabel('Freq(rad/sample)')
plt.grid()

#Question1 for N= 8
h1b = np.concatenate((np.ones(7),h1))
w, h = signal.freqz(h1b,whole = True)
y = w[::-1]
y = -1*y
w = np.concatenate((y,w))
h = np.concatenate((h,h))
fig = plt.figure()
plt.title('frequency response of step function N = 8')
plt.plot(w,h)
plt.ylabel('H(e^jw)')
plt.xlabel('Freq(rad/sample)')
plt.grid()
```

```python
#Question2 for N= 4
x = np.arange(-4,5)
y = 1 - abs(x)/4
w = np.linspace(-2*np.pi,2*np.pi,1024)
res1 = np.zeros((1024))
for i in range(1024):
    for j in range(9):
        res1[i] = res1[i] + y[j]*np.exp(-1j*w[i]*x[j])
fig = plt.figure()
plt.title('frequency response of linear function N = 4')
plt.plot(w, res1)
plt.ylabel('H(e^jw)')
plt.xlabel('Freq(rad/sample)')
plt.grid()

#Question2 for N= 8
x = np.arange(-8,9)
y = 1 - abs(x)/4
w = np.linspace(-2*np.pi,2*np.pi,1024)
res2 = np.zeros((1024))
for i in range(1024):
    for j in range(17):
        res2[i] = res2[i] + y[j]*np.exp(-1j*w[i]*x[j])
fig = plt.figure()
plt.title('frequency response of linear function N = 8')
plt.plot(w, res2)
plt.ylabel('H(e^jw)')
plt.xlabel('Freq(rad/sample)')
plt.grid()
```

```python
#Question3 for N= 4
sinc = np.zeros((1000))
N = 4
for i in range(1,1000):
    sinc[i] = np.sin(np.pi*i/N)/(np.pi*i/N)
w, h = signal.freqz(sinc,whole = True)
y = w[::-1]
y = -1*y
w = np.concatenate((y,w))
h = np.concatenate((h,h))
fig = plt.figure()
plt.title('frequency response of sinc function N = 4')
plt.plot(w, h)
plt.ylabel('H(e^jw)')
plt.xlabel('Freq(rad/sample)')
plt.grid()

#Question3 for N= 8
sinc = np.zeros((1000))
N = 8
for i in range(1,1000):
    sinc[i] = np.sin(np.pi*i/N)/(np.pi*i/N)
w, h = signal.freqz(sinc,whole = True)
y = w[::-1]
y = -1*y
w = np.concatenate((y,w))
h = np.concatenate((h,h))
fig = plt.figure()
plt.title('frequency response of sinc function N = 8')
plt.plot(w, h)
plt.ylabel('H(e^jw)')
plt.xlabel('Freq(rad/sample)')
plt.grid()
```
